```python
import pandas as pd
import matplotlib.pyplot as plt

#Leitura dos arquivos
df1 = pd.read_excel('datasets\Aracaju.xlsx')
df2 = pd.read_excel('datasets\Fortaleza.xlsx')
df3 = pd.read_excel('datasets\\Natal.xlsx')
df4 = pd.read_excel('datasets\Recife.xlsx')
df5 = pd.read_excel('datasets\Salvador.xlsx')

df = pd.concat([df1,df2,df3,df4,df5])
#Quantidade de vezes que cada ID da loja aparece 
df["LojaID"].value_counts(ascending=False)
```

```python
#Gráfico de barras
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```python
#Gráfico de barras horizontais
df["LojaID"].value_counts().plot.barh()
```

```python
df['Receita'] = df['Vendas'].mul(df['Qtde'])
#Gráfico de Pizza
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie()
```

```python
#Vezes que cada cidade aparece
df["Cidade"].value_counts()
```

```python
#Adicionando um título e alterando o nome dos eixos
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
#Alterando a cor
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade", color="red")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
#Alterando o estilo
plt.style.use("ggplot")
```

```python
df.groupby(df['Cidade'])["Qtde"].sum().plot(title = "Total Produtos vendidos x Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```python
df.groupby(df["Cidade"])["Qtde"].sum()
```

```python
#Hisograma
plt.hist(df["Qtde"]);
```

```python
plt.scatter(x=df['Data'].dt.day, y = df["Receita"]);
```

```python
#Salvando em png
df.groupby(df["Data"].dt.month)["Qtde"].sum().plot(marker = "v")
plt.title("Quantidade de produtos vendidos x mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend()
plt.savefig("grafico QTDE x MES.png")
```
